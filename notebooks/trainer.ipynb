```python
from surprise import Dataset
from surprise import Reader
from surprise import SVD
from surprise import accuracy
from surprise.model_selection import train_test_split
import pandas as pd
```

```python
reader = Reader(rating_scale=(0.5, 5))
```

```python
ratings = pd.read_csv('../datasets/movielens/ratings.csv')
ratings.head(1)
```

```python
data = Dataset.load_from_df(
    ratings[['userId', 'movieId', 'rating']],
    reader
)
```

```python
train_set,test_set = train_test_split(data,test_size=0.2,random_state=42)
```

```python
model = SVD(n_factors=100,n_epochs=30,random_state=42)
```

```python
model.fit(train_set)
```

```python
preds = model.test(test_set)
preds[0]
```

```python
model.predict(101,106)
```

```python
print("RMSE :", accuracy.rmse(preds))
print("MAE  :", accuracy.mae(preds))
```

```python
print(model.pu.shape)  # User latent factors
print(model.qi.shape)  # Movie latent factors
```
