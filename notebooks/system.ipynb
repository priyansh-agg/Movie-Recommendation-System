```python
import ast
```

```python
def convert(obj):
    li = []
    for i in ast.literal_eval(obj):
        li.append(i['name'].replace(" ",""))
    return ' '.join(li)
```

```python
movies['genres'] = movies.genres.apply(convert)
movies['keywords'] = movies.keywords.apply(convert)
```

```python
movies.isnull().sum()
```

```python
movies
```

```python
def convert3(obj):
    li = []
    count = 0
    for i in ast.literal_eval(obj):
        if(count>=3): break
        li.append(i['name'].replace(" ",""))
        count +=1
    return ' '.join(li)
```

```python
movies.cast = movies.cast.apply(convert3)
```

```python
movies
```

```python
def convertcrew(obj):
    li = []
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            li.append(i['name'].replace(" ",""))
            break
    return ' '.join(li)
```

```python
movies.crew = movies.crew.apply(convertcrew)
movies
```

```python
movies.overview = movies.overview.apply(lambda x:' '.join(x.split()))
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = [ps.stem(i) for i in text.split()]
    return ' '.join(y)
```

```python
movies.tags = movies.tags.apply(stem)
```

```python
movies = movies[['id','title','tags']]
```

```python
movies.tags = movies.tags.apply(lambda x: x.lower())
movies  
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies.tags).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie,similarity):
    idx = movies[movies['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])[1:6]
    for i in distances:
        print(movies.iloc[i[0]].title)
```

```python
recommend('The Dark Knight',similarity)
```

```python
movie_to_index = pd.Series(movies.index, index=movies['title']).to_dict()
movie_to_index
```

```python
def recommend_multiple(movie_list, similarity,top_n=5):
    indices = [movie_to_index[movie] for movie in movie_list if movie in movie_to_index]

    if not indices:
        return

    avg_similarity = np.mean(similarity[indices], axis=0)
    print(avg_similarity.shape)
    distances = sorted(
        list(enumerate(avg_similarity)),
        reverse=True,
        key=lambda x: x[1]
    )

    count = 0
    for idx, score in distances:
        if idx not in indices:
            print(movies.iloc[idx].title)
            count += 1

        if count == top_n:
            break
```

```python
recommend_multiple(
    ["Interstellar", "Inception", "The Prestige"],
    similarity
)
```
